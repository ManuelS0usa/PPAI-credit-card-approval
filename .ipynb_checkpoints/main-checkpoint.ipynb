<h3> Important Notes </h3>

- The target variable show imbalanced proportion which affects the prediction capability. Although I have trained some models that show a good scoring values, this is misleading due to the fact that confusion matrix and AUC value have poor performance. In order to resolve this issue im going to apply some data cleaning: remove outliers, reduce dimensionality and if necessary, apply some undersampling;

- Years employed and income features show a lot of outliers.


```python
# DBSCAN
from sklearn.cluster import DBSCAN
outlier_detection = DBSCAN(eps=0.5, min_samples=2)
clusters = outlier_detection.fit_predict(df_ori)
print(clusters)

# concatenate clusters with customer data
df_ori['clusters'] = pd.Series(clusters, index=df_ori.index)
print(df_ori)

outliers = df_ori[df_ori['clusters']==-1]
print(len(outliers))
```

```python
df_ori['clusters'].value_counts()
```

```python
df_ori = df_ori[df_ori['clusters'] != -1]
```

```python
df_ori.shape
```

```python
df_ori = df_ori.drop(['clusters'], axis=1)
```

```python
df_ori['target'].value_counts()
```

```python
# target vector - target
y = df_ori['target']

# feature matrix
X = df_ori.loc[:, df_ori.columns != 'target']
```

```python
from sklearn.model_selection import train_test_split

# training and testing splitting
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
print(X_train.shape)
print(X_test.shape)
```

```python
# Standardize the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA

# reduced dataset dimensionality ir order to select best features and improve training results
pca = PCA(n_components=14)
pca.fit(X_train_scaled)
# transform data accordinh to the chosen components
X_pca_train = pca.transform(X_train_scaled)
X_pca_test = pca.transform(X_test_scaled)
print("Original shape: {}".format(str(X_train_scaled.shape)))
print("Reduced shape: {}".format(str(X_pca_train.shape)))
```
